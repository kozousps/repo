Load Dataframe and name Headers

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn')

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data',
                 names = ['classes', 'cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor', 'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
                         'stalk-shape', 'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring',
                         'veil-type', 'veil-color', 'ring-number', 'ring-type', 'spore-print-color', 'population', 'habitat'])
```

Select Desired Series

```python
result = df[['classes', 'odor', 'habitat']]
```

```python
result['classes'] = result['classes'].replace(['e', 'p'], ['edible', 'poisonous'])
classes = result.groupby(['classes']).size()

plt.bar(classes.index, classes)
plt.title('Number of Edible/Poisonous Mushrooms in DataFrame')
plt.show()
```

```python
result['odor'] = result['odor'].replace(['a', 'l', 'c', 'y', 'f', 'm', 'n', 'p', 's'], ['almond', 'anise', 'creosote', 'fish', 'foul', 'musty', 'none', 'pungent', 'spicy'])
odor = result.groupby(['odor']).size()

plt.bar(odor.index, odor)
plt.title('Number of Odorous Mushrooms in DataFrame')
plt.show()
```

```python
result['habitat'] = result['habitat'].replace(['g', 'l', 'm', 'p', 'u', 'w', 'd'], ['grasses', 'leaves', 'meadows', 'paths', 'urban', 'waste', 'woods'])
habitat = result.groupby(['habitat']).size()

plt.bar(habitat.index, habitat)
plt.title('Number of Mushroom Habitats in DataFrame')
plt.show()
```

Format for Stacked Bar Chart

```python
co = result[['classes', 'odor']]
ho = result[['odor', 'habitat']]
hc = result[['classes', 'habitat']]
```

```python
title = 'Charting Mushroom Odors to Toxicity'
co.groupby(['odor', 'classes']).size().unstack().plot(kind='bar', stacked=True, title = title)
```

Most nonscented mushrooms are safe to eat. Almond and anise smelling mushrooms are always good to eat. The rest can be assumed to be poisonous.

```python
title = 'Charting Mushroom Odors to Habitat'
ho.groupby(['odor', 'habitat']).size().unstack().plot(kind='bar', stacked=True, title = title, colormap='Paired')
```

Most mushrooms have no smell, followed by foul smelling mushrooms. 

```python

title = 'Charting Mushroom Toxicity to Habitat'
hc.groupby(['classes', 'habitat']).size().unstack().plot(kind='bar', stacked=True, title = title, colormap='Paired')
```

Edible mushrooms tend to be in grasses, woods, meadows, and waste. Poisonous mushrooms are more likely to be in leaves, urban areas, and paths.
