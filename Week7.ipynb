# Table of Average movie scores by friends

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/kozousps/repo/main/week7.csv')
df
```

Define table shape and averages. 
Append results to new table.

```python
c = df.shape[1]
r = df.shape[0]
apple = ((df.iloc[:].sum(axis=0))[1:])/r
df = df.append(apple, ignore_index=True)
df['Friend Review Average'] = (df.sum(axis=1)/c)
df.loc[df.tail(1).index, 'Unnamed: 0'] = 'Total'
df
```
