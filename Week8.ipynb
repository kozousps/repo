```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn')

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data', sep="\s+", 
                 names=['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name'])
df['horsepower'] = df['horsepower'].replace(['?'], np.nan)
df['horsepower'].apply(pd.to_numeric).astype('Int64')
df['origin'] = df['origin'].replace([1, 2, 3], ['USA', 'Asia', 'Europe'])
```

Create bar char of distribution of cylinders

```python
cylinders = df.groupby(['cylinders']).size()
cylinders.plot.bar()
```

Plotting Horsepower and Weight on Scatter plot

```python
weight = df['weight'].apply(pd.to_numeric).astype('float')
horsepower = df['horsepower'].apply(pd.to_numeric).astype('float')
plt.scatter(weight, horsepower)
plt.title('horsepower vs weight')
plt.xlabel('weight')
plt.ylabel('horsepower')
plt.show()
```

```python
df.groupby(df['model year'])['acceleration'].describe()
```

Plot average acceleration rate by model year

```python
years = df.groupby(df['model year'])['acceleration'].mean()
years.plot.bar()
```
